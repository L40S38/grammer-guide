# TOEIC英文法・時間軸ビジュアル教材

このNotebookは確認用です。YAMLと `src/grammar_guide` のレンダラーを使って、公開ページと同じデータを縦に表示します。

```python
from pathlib import Path

from IPython.display import HTML, Markdown, display

from grammar_guide.load_data import load_all_data
from grammar_guide.render_cards import render_flow_steps, render_quiz_card, render_step_cards
from grammar_guide.render_tables import render_styled_table
from grammar_guide.render_timeline import render_time_axis

PROJECT_ROOT = Path.cwd().parent
if not (PROJECT_ROOT / "data").exists():
    PROJECT_ROOT = Path.cwd()

data = load_all_data(PROJECT_ROOT / "data")
sections = {s["section_id"]: s for s in data["sections"]}
```

```python
def show_section(section_id: int) -> None:
    sec = sections[section_id]
    display(Markdown(f"## {sec['title']}"))
    display(Markdown(sec.get("description", "")))

    if section_id == 1:
        display(HTML(render_styled_table(data["forms_df"], section_id)))
    elif section_id == 2:
        display(HTML(render_time_axis(data["tenses_items"], "時制・完了形の時間軸")))
        display(HTML(render_styled_table(data["tenses_df"], section_id)))
    elif section_id == 3:
        display(HTML(render_time_axis(data["conditionals_items"], "条件文・仮定法の時間軸")))
        display(HTML(render_styled_table(data["conditionals_df"], section_id)))
    elif section_id == 4:
        display(HTML(render_styled_table(data["modals_df"], section_id)))
    elif section_id == 5:
        display(HTML(render_time_axis(data["modal_perfects_items"], "助動詞 + have done の時間軸")))
        display(HTML(render_styled_table(data["modal_perfects_df"], section_id)))
    elif section_id == 6:
        display(HTML(render_time_axis(data["confusions_items"], "混同しやすい形の比較")))
        display(HTML(render_styled_table(data["confusions_df"], section_id)))
    elif section_id == 7:
        display(HTML("<h4>自然な候補</h4>" + render_styled_table(data["toeic_natural_df"], section_id)))
        display(HTML("<h4>不自然になりやすい候補</h4>" + render_styled_table(data["toeic_unnatural_df"], section_id)))
        display(HTML(render_flow_steps(data["toeic_patterns"].get("flow_steps", []))))
    elif section_id == 8:
        display(HTML(render_time_axis(data["decision_table_items"], "判定表の時間軸")))
        display(HTML(render_styled_table(data["decision_table_df"], section_id)))
    elif section_id == 9:
        display(HTML(render_step_cards(data["check_steps"].get("steps", []))))
        quiz = data["check_steps"].get("quiz", {})
        display(HTML(render_quiz_card(quiz.get("sentence", ""), quiz.get("options", []))))

    display(Markdown(f"> {sec.get('point_box', '')}"))
```

```python
for section_id in range(1, 10):
    show_section(section_id)
```
